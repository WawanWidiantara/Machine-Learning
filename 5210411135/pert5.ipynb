## Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read CSV file

```python
data = pd.read_csv('C:/Users/LABKOM/Downloads/5210411135/data/hasil-preprocessimng-data-pert-3.csv')
data.head()
```

```python
# cek type data dan jumlah kolom
print(data.info()) 
```

```python
# cek statistik
print(data.describe())
```

```python
# cek jumlah kemunculan value tiap kolom
data['Wilayah'].value_counts()
```

#### Data Visualization

```python
# with seaborn
sns.set(style='white')
ax = sns.countplot(x='Wilayah', data=data)
```

```python
# with matplot lib
data.Wilayah.value_counts().plot.bar(x='Wilayah')
data.Wilayah.value_counts() 
```

```python
numerik = data.select_dtypes(include=['float64'])
indx = numerik.columns
data.hist(column=indx, figsize=(10,20), layout=(5,1))
plt.show()
```
