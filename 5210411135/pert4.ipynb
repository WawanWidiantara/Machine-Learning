### Pertemuan 4 - Normalisasi

- Normalisasi   -> nilainya 0 - 1 (biasa digunakan pada distribusi normal [bersifat range])
- Standarisasi  -> rata-ratanya = 1 dan standar deviasi = 0

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelBinarizer
```

```python
df = pd.read_csv("C:/Users/LABKOM/Downloads/5210411135/data/hasil-preprocessing-data-pert-3-numeric.csv")

# show dataframe
df
```

```python
df.describe() # describe dataframe (secara statistik)
```

```python
df.loc[2,'Jum_kamar'] = 100 # mengganti value ke 2 dari Jum_kamar
print(df)
print(df.shape) # menampilkan dimensi dataframe
print(df.Jum_kamar)
```

```python
fitur_minmax = df.copy() # copy dataframe ke variable baru
features = fitur_minmax[['Jum_kamar']]
min_max = MinMaxScaler()
kamar_minmax = min_max.fit_transform(features.values) # only work in 2D array

kamar_minmax
```

```python
fitur_minmax[['Jum_kamar']] = kamar_minmax # ganti value kolom jum_kamar sesuai normalisasi

fitur_minmax
```
